```python
%matplotlib inline
import matplotlib.pyplot as plt

##this is going to support Chinese ,but it doesn't work
import matplotlib as mpl 
mpl.rcParams['font.sans-serif']=['SimHei']
mpl.rcParams['axes.unicode_minus']=False
```

```python
labels = 'frog', 'hogs', 'dogs', 'logs'
sizes = 15, 20, 45, 20
colors = 'yellowgreen', 'gold', 'lightskyblue', 'lightcoral'
explode = 0.1, 0.2, 0, 0.1
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=50)
plt.axis('equal')
plt.show()
```

```python
## trying to paint one cos and sin function
import numpy as np

## set the range of x axis
plt.xlim(-4.0,4.0)
## set all points for x axis
plt.xticks(np.linspace(-4,4,21,endpoint=True))

## same to y axis
# plt.ylim(-1.0,1.0)
# plt.yticks(np.linspace(-1,1,5,endpoint=True))

X=np.linspace(-np.pi,np.pi,256,endpoint=True)#https://docs.scipy.org/doc/numpy/reference/generated/numpy.linspace.html
C,S,T=np.cos(X),np.sin(X),np.tan(X)
plt.plot(X,C)
plt.plot(X,S)

plt.show()
```

```python

# plt.plot(X,np.ones(len(X)))
plt.plot(X,T)### what such a  cute img of tan function
plt.show()
```

```python
### we can paint one empty plot

plt.xlim(-4.0,4.0)
## set all points for x axis
plt.xticks(np.linspace(-4,4,21,endpoint=True))

## same to y axis
plt.ylim(-1.0,1.0)
plt.yticks(np.linspace(-1,1,5,endpoint=True))
plt.show()
```

```python
ax = plt.gca()
ax.spines['right'].set_color('none') 
ax.spines['top'].set_color('none') 
ax.xaxis.set_ticks_position('bottom') 
ax.spines['bottom'].set_position(('data',0)) 
ax.yaxis.set_ticks_position('left') 
ax.spines['left'].set_position(('data',0)) 
plt.plot(X, C, color='blue', linewidth=2.5, linestyle='-', label='cos') 
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-", label='sin')
plt.legend(loc='upper left') 

plt.show()
```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
T = np.arctan2(Y,X)

plt.axes([0.025,0.025,0.95,0.95])
plt.scatter(X,Y, s=75, c=T, alpha=.5)

plt.xlim(-1.5,1.5), plt.xticks([])
plt.ylim(-1.5,1.5), plt.yticks([])
# savefig('../figures/scatter_ex.png',dpi=48)
plt.show()
```
